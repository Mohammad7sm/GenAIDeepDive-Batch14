```python
from langchain_ollama import ChatOllama
from langchain_core.prompts import ChatPromptTemplate
```

```python
prompt = ChatPromptTemplate.from_template(
        #input_variables=["person"], 
        "Tell me the key achivements of {person} in four bulleted points, if you don't have information please say I don't know ,please help me with correct responses only"
    )
```

```python
prompt = prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant. Please respond to the user as helpfully and accurately as possible."),
    ("user", "Tell me the key achievements of {person} in four bulleted points. If you don't have information, please say I don't know.")
])
```

```python
llm = ChatOllama(model="llama3.2:latest")
```

```python
chain = prompt | llm
```

```python
response = chain.invoke({"person": "Mahatma Gandhi"})
```

```python
print(response.content)
```
